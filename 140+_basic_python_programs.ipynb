# Program 3
 Write a Python program to find the area of a triangle.

```python
base = eval(input("Enter the base of the triangle in cm: "))
height = eval(input("Enter the height of the triangle in cm: "))
area = 0.5 * base * height
print(f"The area of triangle with {base} base and {height} height is {area} cm^2.")
```

#Program 4
 Write a Python program to swap two variables.

```python
a = int(input("Enter num1: "))
b = int(input("Enter num2: "))
print(f"before swaping {a} and {b}")
temp = a
a = b
b = temp
print(f"after swaping: {a} and {b}")
```

# Program 7
 Write a Python program to convert Celsius to Fahrenheit.

```python
celsius = eval(input("Enter temperature in celsius: "))
Fahrenheit = (9 * celsius / 5 ) + 32
print(Fahrenheit)
```
