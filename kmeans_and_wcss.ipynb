K-MEANS
WCSS = within cluster sum of squares
1. From above scatter plots we observed there might be 5 optimal clusters can be available
2. in order to get optimal cluster we will use elbow method
3. elbow method is a graph between cluster vs WCSS
4. so first we will calculate WCSS for one cluster i.e., k=1
5. then we repeat the process for multiple clusters
6. then we plot the graph



First we will check WCSS for one cluster
